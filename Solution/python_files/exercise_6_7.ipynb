```python
import pandas as pd
```

```python
data_ndvi = [[3,0.84323,0.824573,0.320378],
             [5,0.773826,0.59776,0.661778],
             [48,0.866923,0.6865357,0.764982],
             [50,0.867683,0.828345,0.791617],
             [94,0.835853,0.7451488,0.729657],
             [97,0.8587167,0.727397,0.306645]]
```

```python
df_NDVI = pd.DataFrame(data_ndvi, columns=["Polygon_id","May_NDVI","June_NDVI","July_NDVI"])
```

```python
df_NDVI
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure(dpi= 136, figsize=(8,4))
N = len(df_NDVI['Polygon_id'])
x = np.arange(N)


plt.xticks(x, list(df_NDVI['Polygon_id']))
plt.plot(x,'May_NDVI',  data=df_NDVI, marker='o', markerfacecolor='red',  color='red', linewidth=2)
plt.plot(x,'June_NDVI',  data=df_NDVI, marker='o', markerfacecolor='green',  color='green', linewidth=2)
plt.plot(x, 'July_NDVI',  data=df_NDVI, marker='o', markerfacecolor='blue',  color='blue', linewidth=2)
plt.legend()

plt.xlabel('Polygons of Interest')
plt.ylabel('NDVI')
plt.title('NDVI Analysis')
plt.xticks(rotation=90)
#plt.savefig('C:/Users/Toshiba/Desktop/Master courses/Geoinformatics/Exercises/Exam files/cumulative_prec_Olpe_HSK')


plt.show()


```

```python
ndvi_trans=df_NDVI.transpose()
ndvi_trans.columns=ndvi_trans.iloc[0]
ndvi_trans=ndvi_trans.drop(['Polygon_id'])
ndvi_trans = ndvi_trans.astype(float)
ndvi_trans
```
